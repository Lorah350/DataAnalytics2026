```python
## Lab 3. Employee data
This lab exercise makes updates to information extracted from
an employee database.
```

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
# Update Mark's last name
updated_list = []

for emp in hr_list:
    if emp[2] == 'Mark Blick':
        updated_list.append((emp[0], emp[1], 'Mark Blick-Hawley', emp[3]))
    else:
        updated_list.append(emp)

hr_list = updated_list
```

```python
print(hr_list)
```

```python
Employee# | DeptCode | Name | NN,NNN
```

```python
for emp in hr_list:
    emp_num = emp[0]
    dept = emp[1]
    name = emp[2]
    salary = int(emp[3])  # convert to int so we can format it

    print(f"{emp_num} | {dept} | {name} | {salary:,}")
```

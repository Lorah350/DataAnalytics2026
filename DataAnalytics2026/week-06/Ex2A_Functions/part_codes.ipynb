```python
# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

def get_supplier_code(code):
    # Find the position of the colon
    colon_index = code.find(':')
    # Everything before the colon is the supplier
    return code[:colon_index]
```

```python
def get_product_num(code):
    colon_index = code.find(':')
    dash_index = code.find('-')
    # Slice between colon and dash
    return code[colon_index + 1 : dash_index]
```

```python
def get_size(code):
    dash_index = code.find('-')
    # Everything after the dash is the size
    return code[dash_index + 1 :]
```

```python
code1 = "ACME:123-L"
code2 = "DI:12345-M"
code3 = "ACE:1-12"
```

```python
supplier = get_supplier_code(code1)
product = get_product_num(code1)
size = get_size(code1)

print(f"For {code1}, the supplier is {supplier}. Part number {product} is available in size {size}.")
```

```python
supplier = get_supplier_code(code2)
product = get_product_num(code2)
size = get_size(code2)

print(f"For {code2}, the supplier is {supplier}. Part number {product} is available in size {size}.")
```

```python
supplier = get_supplier_code(code3)
product = get_product_num(code3)
size = get_size(code3)

print(f"For {code3}, the supplier is {supplier}. Part number {product} is available in size {size}.")
```

```python
codes = ["ACME:123-L", "DI:12345-M", "ACE:1-12"]

for c in codes:
    supplier = get_supplier_code(c)
    product = get_product_num(c)
    size = get_size(c)
    print(f"For {c}, the supplier is {supplier}. Part number {product} is available in size {size}.")
```

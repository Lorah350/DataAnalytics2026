## Lab 2 of Exercise 2.B
Working with datetime functions.


```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
today
```

```python
today_str = today.strftime("%A, %B %d, %Y")
print(f"Today is {today_str}")
```

# My birthday


```python

my_birthday = datetime.date(2026, 5, 12)
my_birthday
```

```python
birthday_str = my_birthday.strftime("%m/%d/%Y")
print(f"My birthday is {birthday_str}")
```

# Next 90 Days


```python
ninety_d = today + datetime.timedelta(days=90)
ninety_d
ninety_d_str = ninety_d.strftime("%B %d, %Y")
print(f"90 days from today is {ninety_d_str}")
```

# Time for Dinner


```python
dinner_time = datetime.time(18, 30)
dinner_time

dinner_str = dinner_time.strftime("%I:%M %p")
print(f"Let's meet for dinner at {dinner_str}")
```

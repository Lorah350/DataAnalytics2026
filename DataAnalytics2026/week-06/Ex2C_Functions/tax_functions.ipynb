## Creating Tax Functions

In this lab, I’m practicing how to define functions that calculate different types of taxes based on gross pay. I created three functions: one for Social Security tax, one for Medicare tax, and one for federal tax using a withholding code. Then I tested each function with different examples to make sure the calculations work correctly.


```python
# Social Security tax is 6.2% of gross pay

def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

```python
# Medicare tax is 1.45% of gross pay

def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

```python
# Federal tax depends on the withholding code.
# Codes 0–3 have fixed rates.
# Code 4 and above reduces the rate by 0.5% for each number over 4.

def get_federal_tax(gross_pay, withholding_code):
    
    if withholding_code == 0:
        rate = 0.23
    elif withholding_code == 1:
        rate = 0.21
    elif withholding_code == 2:
        rate = 0.195
    elif withholding_code == 3:
        rate = 0.185
    else:
        # Code 4+ → 18% minus 0.5% for each number over 4
        rate = 0.18 - ((withholding_code - 4) * 0.005)
    
    return gross_pay * rate
```

```python
gross = 750
code = 0

print("Person 1:")
print("Social Security Tax:", get_soc_sec_tax(gross))
print("Medicare Tax:", get_medicare_tax(gross))
print("Federal Tax:", get_federal_tax(gross, code))
```

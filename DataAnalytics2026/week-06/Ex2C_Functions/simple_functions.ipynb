## Creating and Calling Simple Functions

In this lab, I’m practicing how to define my own custom functions in Python.  
Each function will take in some information and return a sentence or message based on the inputs.  
This helps me understand how parameters and return values work in real code.


```python
# This function returns three lines about my favorite things.
# I pass in my name, my favorite movie, and my favorite musician/band.

def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician/band is {music}."

# Calling the function with my own info
print(favorite_things("Lorah", "The Woman King", "Beyoncé"))
```

```python
# This function returns a sentence explaining why I'm in the Year Up program.

def why_im_here(purpose):
    return purpose

# Calling the function with my reason
print(why_im_here("I joined Year Up to build my skills in data analytics and open better career opportunities."))
```

```python
# This function describes one of my favorite places and why I love it.

def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place("Victoria Falls", "the view is beautiful and it feels peaceful"))
```

```python
print(favorite_place("Miami Beach", "I love the warm weather and the ocean"))
```

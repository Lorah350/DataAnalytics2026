```python
import numpy as np
```

## Array Basics
Below are examples of creating Python lists and converting them into NumPy arrays.


```python
fruits = ["apple", "banana", "cherry", "date"]
fruits
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
arr_1.dtype
```

The dtype tells us the data type stored in the array.
For example, dtype('<U6') means:
- U = Unicode string
- 6 = maximum length of the strings in the array


```python
arr_2 = np.array([
    [10, 11, 12],
    [13, 14, 15]
])

print(arr_2)
```

```python
print(arr_2.T)
```

Using arr_2.T does NOT change the original array.
It only returns a transposed *view* of the data.
arr_2 itself stays the same.


```python
arr_3 = np.array([
    [1, 3, 5],
    [2, 4, 6]
])

print(arr_3)
```

```python
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```python
arr_5 = np.ones((2, 2, 2))
print(arr_5)
```

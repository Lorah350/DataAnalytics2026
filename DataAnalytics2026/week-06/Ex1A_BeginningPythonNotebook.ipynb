```python
# What other number or string value(s) can be typecast as boolean values? Use the print() function to list them
print(bool(0))        # False
print(bool(""))       # False (empty string)
print(bool(None))     # False
# These are the only things that python treats as false

print(bool(1))        # True
print(bool(-5))       # True
print(bool(100))      # True
print(bool("hello"))  # True
print(bool("0"))      # True (non-empty string)
print(bool("False"))  # True (still non-empty)
# anything non-zero or not empty records a true
```

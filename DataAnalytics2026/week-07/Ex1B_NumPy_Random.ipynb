# Exercise 1.B NumPy Random

This exercise explores generating random numbers, random selections, weighted choices, and normal distributions using NumPy.


```python
import numpy as np
```

## Generating Random Numbers


```python
r_1 = np.random.randint(1, 21, size=2)
print(r_1)
```

```python
r_2 = np.random.randint(1, 21, size=(1, 2))
print(r_2)
```

```python
r_3 = np.random.rand(2, 1)
print(r_3)
```

```python
r_4 = np.random.randint(20, 101, size=(4, 5))
print(r_4)
```

```python
r_5 = np.random.rand(4, 5)
print(r_5)
```

## Generating Random Choices from an Array


```python
students = np.array([
    "Ava", "Ben", "Carlos", "Diana",
    "Eli", "Fatima", "Grace", "Hector",
    "Ivy", "Jamal", "Kara", "Liam"
])
students
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4), replace=False)
print(breakout_groups)

# Comment:
# This is NOT an effective way to assign breakout groups because the sampling is random
# and does not guarantee balanced groups or prevent repeating the same groupings each run.
```

```python
# Example probability distribution (must sum to 1)
p_values = [
    0.00,  # Ava (never selected)
    0.05,  # Ben
    0.10,  # Carlos
    0.05,  # Diana
    0.10,  # Eli
    0.30,  # Fatima (30% chance)
    0.10,  # Grace
    0.10,  # Hector
    0.05,  # Ivy
    0.05,  # Jamal
    0.05,  # Kara
    0.05   # Liam
]

call_on_weighted = np.random.choice(students, size=5, p=p_values)
print(call_on_weighted)

# Comment:
# Fatima has the highest chance (30%).
# Ava has 0% chance and will never be selected.
# Other students have smaller but non-zero probabilities.
```

## Generating a Normal Distribution


```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```

## Permutations


```python
print("Original r_4:")
print(r_4)
```

```python
np.random.shuffle(r_4)
print("Shuffled r_4:")
print(r_4)
```

Shuffling modifies the array *in place*. NumPy rearranged the rows of r_4, but did not return a new array. It’s like shuffling a deck — the deck itself changes.


```python
new_shuffle = np.random.permutation(r_4)
print("Permutation result:")
print(new_shuffle)

print("r_4 after permutation (unchanged):")
print(r_4)
```

Shuffle changes the original array directly — like rearranging furniture in your room.

Permutation returns a *new* rearranged array and leaves the original untouched — like taking a picture of your room and rearranging the picture instead.


## Mathematical Operations


```python
print("r_1:\n", r_1)
print("shape:", r_1.shape)

print("r_2:\n", r_2)
print("shape:", r_2.shape)

result = np.add(r_1, r_2)
print("r_1 + r_2:\n", result)
print("shape:", result.shape)
```

```python
print("r_1 shape:", r_1.shape)
print("r_3 shape:", r_3.shape)

result = np.add(r_1, r_3)
print(result)
print("shape:", result.shape)

# COMMENT:
# This error is expected. r_1 has shape (2,3) and r_3 has shape (4,5).
# These shapes are NOT compatible for broadcasting because:
# - None of the dimensions match
# - Neither dimension is 1 (so NumPy cannot stretch either array)
# Because of this, NumPy cannot align the arrays, so it raises a ValueError.
```

```python
print("r_2 shape:", r_2.shape)
print("r_4 shape:", r_4.shape)

result = np.add(r_2, r_4)
print(result)
print("shape:", result.shape)
```

```python
# r_1 + r_4
try:
    print(np.add(r_1, r_4))
except Exception as e:
    print("Error for r_1 + r_4:", e)

# r_2 + r_4
try:
    print(np.add(r_2, r_4))
except Exception as e:
    print("Error for r_2 + r_4:", e)

# r_3 + r_4
try:
    print(np.add(r_3, r_4))
except Exception as e:
    print("Error for r_3 + r_4:", e)

# Comment:
# These errors happen because the shapes are not compatible for broadcasting.
```

```python
print(r_4 + np.array([1]))
print(r_4 + np.array([[1]]))
print(r_4 + np.array([1,1,1,1,1]))

try:
    print(r_4 + np.array([[1],[1]]))
except Exception as e:
    print("Error:", e)
```

Broadcasting works when NumPy can stretch one array to match the shape of another.

Using add(), each element of the smaller array gets added to the matching position in the larger array. NumPy compares shapes from right to left — if a dimension is 1 or matches, it can broadcast. The resulting array always takes the larger shape.

## Grocery Problem


```python
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)
```

```python
total1 = np.sum(bag1)
total2 = np.sum(bag2)

print(f"The first bag contains {total1} items, and the second bag contains {total2} items.")
```

```python
combined = bag1 + bag2
print(f"There are {combined[0]} vegetables, {combined[1]} fruits, {combined[2]} boxes of snacks, and {combined[3]} beverages in the two grocery bags.")
```

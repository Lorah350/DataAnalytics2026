# Exercise 3.A Getting familiar with Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt

# Bringing in numpy for data handling
# Bringing in matplotlib so I can start plotting graphs
```

## Creating Basic Plots


```python
# Setting up my x and y values for the plot
# These are simple numbers just to practice plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
# First basic plot just to see the line graph
plt.plot(x, y)
plt.show()
```

```python
# Same plot but now I'm adding markers to show each point clearly
plt.plot(x, y, marker='o')
plt.show()
```

```python
# Adding a title and labeling the axes to make the graph more readable
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Styling Plots


```python
# Trying out a dashed line style and changing the color to red
plt.plot(x, y, linestyle='--', color='red')
plt.show()
```

```python
# Changing marker style to squares, making them bigger, and coloring them green
plt.plot(x, y, marker='s', markersize=10, markerfacecolor='green')
plt.show()
```

```python
# Experiment 1: Triangle markers with a purple line
plt.plot(x, y, marker='^', color='purple')
plt.show()
```

```python
# Experiment 2: Star markers with an orange line
plt.plot(x, y, marker='*', color='orange')
plt.show()
```

# Exercise 3.B More plotting practice


```python
# Just making sure numpy and matplotlib are loaded again
# (especially if I restarted the notebook)
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Advanced Plotting Techniques
```

```python
# Setting up product names and their sales numbers
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

# Basic bar chart to visualize product sales
plt.bar(products, sales)
plt.show()
```

```python
# Customizing my bar chart to make it look cleaner and more readable
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
# Creating random x and y values to practice scatter plotting
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
# Adding marker sizes and a colormap to make the scatter plot more interesting
sizes = np.random.rand(50) * 100  # marker sizes
colors = np.random.rand(50)       # color values

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis')
plt.title('Customized Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.colorbar()  # adds a color scale on the side
plt.show()
```

## Subplots


```python
# Creating a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- First subplot: Sine wave ---
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax1.plot(x, y)
ax1.set_title('Sine Wave')
ax1.set_xlabel('X values')
ax1.set_ylabel('sin(x)')

# --- Second subplot: Histogram ---
data = np.random.randn(1000)

ax2.hist(data, bins=30, color='skyblue')
ax2.set_title('Histogram of Random Data')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

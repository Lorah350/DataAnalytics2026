# Exercise 1.A NumPy: Manipulating Arrays

This exercise explores looping through arrays, joining and stacking arrays, splitting arrays, and searching/filtering values using NumPy.


```python
import numpy as np
```

## Looping Through Arrays


```python
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])

print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

The first loop prints each *row* multiplied by 2, so the output is in 2D chunks.
The second loop prints each *individual element* multiplied by 2, so the output is 1D values printed one at a time.


```python
for x in arrA:
    loop_result = x * 2
    print(loop_result)

# Comment:
# This prints each entire row multiplied by 2 at once because x is the whole row,
# not a single element. That's why the output is grouped in arrays instead of single numbers.
```

## Joining and Stacking Arrays


```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
# This will cause an error because joined_array_1 and joined_array_2
# do NOT have the same shape.
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)

# Comment:
# The error says the input arrays must have the same shape.
# joined_array_1 and joined_array_2 have different dimensions, so stacking fails.
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA), axis=0)
print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
print(another_2)

# Comment:
# Error: the arrays cannot be concatenated along axis=1 because their row lengths differ.
# NumPy requires matching dimensions along all other axes.
```

```python
h = np.hstack((arrA, arrA))
print(h)
# hstack places arrays side-by-side (horizontal stacking).

v = np.vstack((arrA, arrA))
print(v)
# vstack places arrays on top of each other (vertical stacking).

d = np.dstack((arrA, arrA))
print(d)
# dstack stacks arrays along a new 3rd dimension.
```

## Splitting Arrays


```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)

# Comment:
# Result: 2 arrays, each 2D, each containing half of joined_array_1.
```

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)

# Comment:
# Result: 3 arrays. NumPy splits as evenly as possible.
# Some arrays may have different row counts.
```

## Searching and Replacing Array Values


```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
```

## Sorting and Filtering Arrays


```python
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])

arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)

z = arrB < 7
arrC = arrB[z]
print(arrC)
```

```python
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```

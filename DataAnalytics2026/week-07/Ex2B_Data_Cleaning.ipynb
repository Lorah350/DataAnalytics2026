## Modifying a DataFrame

Now I’m cleaning this dataset like it’s Sunday reset day — removing bad rows, fixing missing values, and making everything look organized.


```python
df_cleaned = df.drop(df[df["Score"] < 0].index)
df_cleaned
```

```python
df_cleaned = df_cleaned.drop(columns=["City"])
df_cleaned
```

```python
mean_age = df_cleaned["Age"].mean()
df_cleaned["Age"] = df_cleaned["Age"].fillna(mean_age)
df_cleaned
```

```python
mean_score = df_cleaned["Score"].mean()
df_cleaned["Score"] = df_cleaned["Score"].replace(0, mean_score)
df_cleaned
```
